```python
#print(swiss_df['Location'])

# "Location"
discrete_column = swiss_df['Location']
# frequency
value_counts = discrete_column.value_counts()

# Побудова графіка функції ймовірностей
plt.figure(figsize=(20, 6))
plt.bar(value_counts.index, value_counts.values / len(discrete_column))
plt.xlabel("Значення")
plt.ylabel("Ймовірність")
plt.title("Графік функції ймовірностей для показника 'Location'")
plt.xticks(rotation=45)
plt.show()


_swiss_df = swiss_df.drop(columns=['Location'])
numeric_columns = _swiss_df.columns

#width of classes
N = len(swiss_df['Location'])
M = 1 + 3.32 * math.log10(N)

M = int(M)

if(M < 5):
    M = 5

# Hist KDE
for column in numeric_columns:
    data = _swiss_df[column].sort_values(ascending = True)
    
    plt.figure(figsize=(8, 6))
    sns.histplot(data, bins=M, kde=True)
    plt.title(f'Гістограма з KDE показника "{column}"')
    plt.xlabel('Значення')
    plt.ylabel('Частота / Щільність')
    plt.show()
```

# Ідентифікація норм. розп. за коеф. A та коеф. E

```python
import numpy as np
import scipy.stats as stats

# Визначення списку числових стовпців
numeric_columns = swiss_df.columns[1:]  # Виключаємо перший стовпець з назвами місць
alpha = 0.05  # Рівень значущості

for column in numeric_columns:
    data = swiss_df[column]
    
    skew_test = stats.skewtest(data)[0]
    kurt_test = stats.kurtosistest(data)[0]
    u_95p = 1.96
    
    print(f'Показник "{column}":')
    print(f'Коефіцієнт асиметрії: {skew_test}')
    print(f'Коефіцієнт ексцесу: {kurt_test}')
    
    if(abs(skew_test) < u_95p and abs(kurt_test) < u_95p):
        print('H0 не відхиляється')
    else:
        print('H0 відхиляється')
    
#     # Підрахунок коефіцієнта асиметрії
#     skewness = stats.skew(data)
    
#     # Підрахунок коефіцієнта ексцесу
#     kurtosis = stats.kurtosis(data)
    
    
#     skewness_p_value = stats.normaltest(data).pvalue
#     kurtosis_p_value = stats.normaltest(data).pvalue
    
#     print(f'pvalue: {stats.normaltest(data).pvalue}')
    
#     if skewness_p_value < alpha:
#         print(f'Гіпотеза про рівність нулю асиметрії відхиляється (p-value = {skewness_p_value})')
#     else:
#         print('Гіпотеза про рівність нулю асиметрії не відхиляється')
    
#     if kurtosis_p_value < alpha:
#         print(f'Гіпотеза про рівність нулю ексцесу відхиляється (p-value = {kurtosis_p_value})')
#     else:
#         print('Гіпотеза про рівність нулю ексцесу не відхиляється')
    
    print()

    #P-значення вказує на ймовірність отримати такі або більш віддалені від нормального розподілу дані випадково, якщо дані справді мають нормальний розподіл.
```

# Критерій Шапіро-Уілка

```python
from scipy.stats import shapiro

numeric_columns = swiss_df.columns[1:]
alpha = 0.05

for column in numeric_columns:
    data = _swiss_df[column]
    stat, p = shapiro(data) #кортеж

    
    if p > alpha:
        print(f'Показник "{column}" має нормальний розподіл (p = {p})')
        print(f'Якщо розподіл нормальний то stat ~ 1; (stat = {stat})')
        print()
    else:
        print(f'Показник "{column}" не має нормальний розподіл (p = {p})')
        
        # Логарифмуємо дані
        logged_data = np.log(data)

        plt.figure(figsize=(8, 6))
        sns.histplot(logged_data, bins=M, kde=True)
        plt.title(f'Гістограма з KDE показника "{column}"')
        plt.xlabel('Значення')
        plt.ylabel('Частота / Щільність')
        plt.show()
        
        stats.probplot(logged_data, dist=stats.norm, plot=plt)
        plt.title(f'Ймов. папір {column}')
        plt.show()
        
        # Повторно перевіряємо розподіл логарифмованих даних
        stat, p = shapiro(logged_data)
        
        if p > alpha:
            print(f'Після логарифмування, показник "{column}" має нормальний розподіл (p = {p})')
            print()
        else:
            print(f'Після логарифмування, показник "{column}" не має нормальний розподіл (p = {p})')
            print()
```

# Відновлення розподілів

```python
import scipy.stats as stats
from statsmodels.distributions.empirical_distribution import ECDF

numeric_columns = swiss_df.columns[1:]
alpha = 0.05
distributions = ['norm', 'rayleigh', 'expon']

result_data = []

for column in numeric_columns:
    
    data = swiss_df[column]
    
    best_distribution = None
    best_params = None
    best_ks_statistic = float('inf')
    
    for dist_name  in distributions:

        dist = getattr(stats, dist_name)
        params = dist.fit(data) #значення пар. розп.

        ks_statistic, p_value = stats.kstest(data, dist_name, args=params)
        
        #ks_statistic, p_value = stats.kstest(data, dist_name)
        result_data.append({'Column': column, 'Distribution': dist_name , 'p-Value': p_value})
        
        if ks_statistic < best_ks_statistic:
            best_distribution = dist
            best_params = params
            best_ks_statistic = ks_statistic
        
result_table = pd.DataFrame(result_data)

display(result_table)

print(f"Найкращий розподіл для {column}: {best_distribution.name}")
print(f"Оцінки параметрів: {best_params}")


for column in numeric_columns:
    data = swiss_df[column]

    best_distribution = None
    best_params = None
    best_ks_statistic = float('inf')

    for dist_name in distributions:
        dist = getattr(stats, dist_name)
        params = dist.fit(data)
        ks_statistic, p_value = stats.kstest(data, dist_name, args=params)

        if ks_statistic < best_ks_statistic:
            best_distribution = dist
            best_params = params
            best_ks_statistic = ks_statistic

    print(f"Найкращий розподіл для {column}: {best_distribution.name}")
    print(f"Оцінки параметрів: {best_params}")

    data_sorted = data.sort_values(ascending=True)

    plt.figure(figsize=(15, 5))

    # Гистограмма и PDF
    plt.subplot(131)
    plt.hist(data_sorted, bins=M, density=True, alpha=0.6, color='b', label='Гістограма')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = best_distribution.pdf(x, *best_params)
    plt.plot(x, p, 'k', linewidth=2, label='PDF')
    plt.title('PDF та Histogram')

    # CDF
    plt.subplot(132)
    cdf = best_distribution.cdf(x, *best_params)
    plt.plot(x, cdf, 'r--', label='CDF')
    plt.title('CDF та емпірична CDF')
    #ECDF
    ecdf = ECDF(data_sorted)             # будуємо ЕФР
    vr = np.sort(data_sorted.unique())
    FN = ecdf(vr)              # обчислюємо її значення в точках варіаційного ряду
    plt.step(vr, FN, where='post', color='k')
    plt.plot(vr, FN, 'ok');

    # Q-Q plot
    plt.subplot(133)
    stats.probplot(data_sorted, dist=best_distribution, sparams=best_params, plot=plt)
    plt.title('Ймовірнісний папір')

    plt.suptitle(f'Аналіз {column}')
    plt.show()
```
